```python
print(tree.EmployeeCount.unique())
print(tree.EmployeeNumber.unique())
print(tree.Over18.unique())
print(tree.StandardHours.unique())
```

```python
tree=tree.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis=1)
```

```python
tree.head()
```

```python
tree['Attrition']=np.where(tree['Attrition']=='Yes',1,0)
```

```python
tree_dummies=pd.get_dummies(tree[['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus',
                                  'OverTime']])
```

```python
tree_dummies.head()
```

```python
tree_dummies.shape
```

```python
tree_duplicate=tree.copy()
```

```python
tree=tree.drop(['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus',
                                  'OverTime'],axis=1)
```

```python
tree=pd.concat([tree,tree_dummies],axis=1)
```

```python
tree.head()
```

```python
x=tree.drop(['Attrition'],axis=1)
y=tree['Attrition']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test= train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
X_train
```

# Model-1

```python
from sklearn import tree
dt1=tree.DecisionTreeClassifier()
dt1.fit(X_train,y_train)
y_pred1=dt1.predict(X_test)
```

# Model-2

```python
dt2=tree.DecisionTreeClassifier(criterion='entropy',max_depth=4)
dt2.fit(X_train,y_train)
y_pred2=dt2.predict(X_test)
```

# Model-3

```python
dt3=tree.DecisionTreeClassifier(criterion='entropy',max_depth=2,min_samples_leaf=2)
dt3.fit(X_train,y_train)
y_pred3=dt3.predict(X_test)
```

```python
from sklearn import metrics
print('Accuracy M1:',metrics.accuracy_score(y_test,y_pred1))
print('Accuracy M2:',metrics.accuracy_score(y_test,y_pred2))
print('Accuracy M3:',metrics.accuracy_score(y_test,y_pred3))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={"max_depth":[2,3,4,5],
            "max_features":[1,3,5,10],
            "min_samples_split":[2,3,5,10],
            "criterion":["gini","entropy"],
            "min_samples_leaf":[1,2,3,4]}
grid_search=GridSearchCV(dt1,param_grid,cv=5)
grid_search.fit(X_train,y_train)
grid_search.best_params_
```

# Model-4

```python
dt4=tree.DecisionTreeClassifier(criterion='entropy',max_depth=3,min_samples_leaf=3,max_features=5,min_samples_split=3)
dt4.fit(X_train,y_train)
y_pred4=dt4.predict(X_test)
```

```python
print('Accuracy M4:',metrics.accuracy_score(y_test,y_pred4))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
print("confusion Matrix",confusion_matrix(y_test,y_pred4))
```

```python
print(classification_report(y_test,y_pred4))
```
